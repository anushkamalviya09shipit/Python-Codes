```python
import pandas as pd
df = pd.read_excel('/content/hospital_readmissions_dataset.xlsx')
```

```python
df.head()
```

```python
df.shape
```

```python
df.drop_duplicates(subset = ['Patient ID'], inplace = True)
```

```python
df.columns
```

```python
numeric_cols = ['time_in_hospital', 'n_lab_procedures',
       'n_procedures', 'n_medications', 'n_outpatient', 'n_inpatient',
       'n_emergency']
for col in numeric_cols:
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3 - Q1
  Lower_bound = Q1 - 1.5 * IQR
  Upper_bound = Q3 + 1.5 * IQR
  df = df[(df[col] >= Lower_bound) & (df[col] <= Upper_bound)]
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df['medical_specialty'].unique()
```

```python
df = pd.get_dummies(df, columns = ['medical_specialty','glucose_test',
       'A1Ctest', 'change', 'diabetes_med', 'readmitted'])
```

```python
X = df.drop(columns = ['Patient ID', 'readmitted_no', 'readmitted_yes','age','diag_1', 'diag_2'])
Y = df['readmitted_yes']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score , precision_score, recall_score,f1_score,confusion_matrix
print("Accuracy Score:",accuracy_score(Y_test, Y_pred))
print("Precision Score:",  precision_score(Y_test, Y_pred))
print("Recall Score:",  recall_score(Y_test, Y_pred))
print("F1 Score:",  f1_score(Y_test, Y_pred))
```

```python
numeric_cols1 = ['Age','Heart rate','Previous Hospitalizations','Length of Stay in days']
for col in numeric_cols1:
  Q1 = df1[col].quantile(0.25)
  Q3 = df1[col].quantile(0.75)
  IQR = Q3 - Q1
  Lower_bound = Q1 - 1.5 * IQR
  Upper_bound = Q3 + 1.5 * IQR
  df1 = df1[(df1[col] >= Lower_bound) & (df1[col] <= Upper_bound)]
```

```python
df1.dropna
```

```python
df1['Gender'].unique()
```

```python
df1['Gender'].replace({'female': 'Female','m': 'Male'}, inplace = True)
```

```python
df1 = pd.get_dummies(df1, columns = ['Gender', 'Primary Diagnosis',
                                    'Comorbidities','Surgical Procedures','Severity of Illness'])
```

```python
df1.columns
```

```python
x = df1.drop(columns = ['Patient ID', 'BP','Socioeconomic Status',
                        'Discharge Destination','Length of Stay in days','Room Type'])
y = df1['Length of Stay in days']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(x_train, y_train)
y_pred = linreg.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
print("R2 score: ",r2_score(y_test,y_pred))
```
